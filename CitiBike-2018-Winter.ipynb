### Function for geopy

```python
# create a function to retrieve a list element
def get_start(start_coord_list, position):
    return start_coord_list[position]

def get_end(end_coord_list, position):
    return end_coord_list[position]

def get_distance():
    return geopy.distance.vincenty(get_start(start_coord_list,i), \
                                   get_end(end_coord_list,i)).miles
```

```python
# Test code for geopy
start = get_start(start_coord_list, 1)
end = get_end(end_coord_list, 1)
```

```python
geopy.distance.vincenty(start, end).miles
```

```python
# Append distance to an empty list
distance_list = []

# Iterate thru get_distance to return the distance between coordinates
for i in range(0, len_start_coord_list):
    distance_list.append(get_distance())
```

```python
# Check to confirm distance was properly added to our list
print(len(distance_list))
station_distance.shape
```

```python
# Make distance_list into a series
# Set to var 'distance2' which will become a new col 'distance' added to df
distance = pd.Series(distance_list)
```

```python
# Insert distance2 into df as a col
# loc=11 sets the location at index11 after end station longitude
# value=distance2 sets values of the column to the pd series list created above
station_distance.insert(loc=11, column='Distance(Miles)', value=distance)
```

```python
# Function to convert seconds to minutes
seconds = station_distance['Trip Duration'].tolist()
seconds_len = (len(seconds))
minutes_list = []

def get_min():
    return seconds[i] / 60
```

```python
# Check the length of trip duration list we assigned to seconds, then seconds_len
seconds_len
```

```python
for i in range(0, seconds_len):
    minutes_list.append(get_min())
```

```python
print(len(minutes_list))
```

```python
# Convert minutes_list into a series
# set to 'tripduration_minutes' which will be new column 'tripduration_minutes' added to df
tripduration_minutes = pd.Series(minutes_list)
```

```python
# Insert tripduration_minutes into df as a column
# loc=1 will set location at index1
# value=tripduration_minutes - sets values of the column to pd series list made above

station_distance.insert(loc=1, column='Trip Duration(Minutes)', value=tripduration_minutes)
```
